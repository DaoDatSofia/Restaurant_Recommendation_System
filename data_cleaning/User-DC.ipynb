```python
import pandas as pd
import numpy as np
```

```python
data_user = pd.read_csv('D:\\Workspace\\Data_Mining\\Restaurant_Recomendation_System\\data\\users.csv',index_col = 0)
data_user.head()
data_review = pd.read_csv('D:\\Workspace\\Data_Mining\\Restaurant_Recomendation_System\\data\\IL_review.csv',index_col = 0)
data_review.head()
```

```python
unique_IL_users = data_review.user_id.unique()
user_idx = [i for i in range(data_user.shape[0]) if data_user.user_id[i] in unique_IL_users]
user_df = data_user.iloc[user_idx,:]
```

```python
user_df.index = range(user_df.shape[0])
user_df.head()
```

```python
user_df.isnull().sum()
```

```python
user_df['is_elite'] = ~np.array(user_df.elite.isnull())
user_df.is_elite.describe()
```

```python
len(user_df.yelping_since.unique()) 
```

```python
yr_ls = [eval(x.split('-')[0]) for x in user_df.yelping_since]
user_df['year'] = yr_ls
user_df
```
